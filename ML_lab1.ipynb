We will now take a look at how the "women and children first" policy was respected (or not) on board the RMS Titanic given that there was only a limited number of lifeboat. To do so, we first create a "Status" column to filter the passengers into 4 categories "adult male", "adult female", "female child" and "male child". Please note that we make the choice of considering passengers of unknown age as adults.

```python
titanic_data['Status'] = titanic_data.loc[:, 'Sex']
titanic_data.loc[ titanic_data["Sex"]=="male" , 'Status' ]="adult male"
titanic_data.loc[ titanic_data["Sex"]=="female" , 'Status' ]="adult female"
titanic_data.loc[ (titanic_data["Age"]<18) & (titanic_data["Sex"]=="female") , 'Status' ]="female child"
titanic_data.loc[ (titanic_data["Age"]<18) & (titanic_data["Sex"]=="male") , 'Status' ]="male child"
titanic_data
```

## Part B : Machine Learning

In this section and using simple ML methods, we will try to predict who survives and who dies based on the various known informations about the passengers.

We begin by importing some necessary libraries for decision trees and Machine Learning.

```python
from IPython.display import Image
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
from sklearn.metrics import confusion_matrix, accuracy_score
```

Unfortunately, decision trees in Python only handle numerical data. Therefore, our "Status" variable cannot be used, and we must consider creating the 2 following columns instead :
- A variable "Male" whose value is 1 for male passengers and 0 for female
- A variable "Kid" whose value is 1 for children and 0 otherwise

#### Question 1 : Create these two variables :

```python
titanic_data['Male']=titanic_data.loc[:,'Sex']
titanic_data['Male'] = titanic_data['Male'].map({'male': 1, 'female': 0})
titanic_data['Kid']=titanic_data.loc[:,'Status']
titanic_data['Kid'] = titanic_data['Kid'].map({'male child': 1, 'female child': 1, 'adult male': 0, 'adult female': 0})
titanic_data
```

#### Question 2 : What does the code below do ? Explain the role of each variable that the "train_test_split" function creates as well as the different parameters.

```python
target_labels=titanic_data.Survived
train_predictor_columns = ['Pclass', 'Fare', 'Male', 'Kid'] #we will predict from the variables
train_feats = titanic_data[train_predictor_columns]

x_train,x_test,y_train,y_test  = train_test_split(train_feats,target_labels, test_size=0.3, random_state = 0)
```

<u>Answer 2 :</u>

*Ce code divise les données en deux groupes distincts pour permettre à l'intelligence artificielle d'apprendre sur une partie (80 %) et d'être évaluée sur l'autre (20 %), garantissant ainsi qu'elle est capable de généraliser et non de simplement apprendre les réponses par cœur.
test_size=0.2 : Définit la répartition. 20% des données sont réservées pour le test et 80% pour l'entraînement.
random_state=0 : Fixe le mélange aléatoire. Cela garantit que vous obtiendrez toujours le même découpage à chaque exécution du code, ce qui permet de reproduire vos résultats.*

We will now train a decision tree with a depth of 3

```python
clf =tree.DecisionTreeClassifier(max_depth=3) #On peut jouer sur la profondeur de l'arbre ici
clf=clf.fit(x_train, y_train)
```

#### Question 3 : Using the "classification_report" function from "metrics" with the right parameters, display the metrics associated with this decision tree and comment. In particular, explain what are the different metrics proposed.

```python
from sklearn.metrics import classification_report
# On regarde si l'IA a bien compris ses exercices de révision
y_pred_train = clf.predict(x_train)
print(classification_report(y_train, y_pred_train))
```

<u>Answer 3</u>

*Le modèle présente une performance globale satisfaisante (78 %), mais il est conservateur. Avec une profondeur limitée à 3, l'arbre a probablement identifié des règles simples et fortes pour les décès (ex: Hommes en 3ème classe), ce qui explique le haut score de rappel sur la classe 0. Cependant, il manque de nuances pour capturer la complexité des profils de survivants, d'où un rappel plus faible sur la classe 1. C'est un exemple typique de modèle qui privilégie la simplicité (évitant l'overfitting) au détriment d'une détection exhaustive des cas plus rares (les survivants).*

The code below will display the decision tree (you might have to adjust the figsize to match with your screen resolution) :

```python
plt.figure(figsize=(12,12))
tree.plot_tree(clf,feature_names=train_predictor_columns,class_names=["Survived","Died"])
```

#### Question 5 : Translate this decision tree into simple clauses. Is it coherent ?

<u>Answer 5:</u>

*Règle 1 (Femmes de classes supérieures) :
SI le passager est une Femme (Male <= 0.5) ET est en 1ère ou 2ème classe (Pclass <= 2.5), ALORS elle Survit (112+11 survivantes contre seulement 4+2 décès).

Règle 2 (Femmes de 3ème classe "tarif bas") :
SI le passager est une Femme ET est en 3ème classe ET a payé un billet très peu cher (Fare <= 8.04), ALORS elle Survit (31 survivantes contre 8 décès).

Règle 3 (Femmes de 3ème classe "tarif élevé") :
SI le passager est une Femme ET est en 3ème classe ET a payé un billet plus cher (Fare > 8.04), ALORS elle Meurt (42 décès contre 38 survivantes).

Règle 4 (Hommes au tarif standard) :
SI le passager est un Homme (Male > 0.5) ET a payé un billet inférieur à 26.2$, ALORS il Meurt (Peu importe l'âge, 281+15 décès contre seulement 37+11 survivants).

Règle 5 (Hommes au tarif élevé) :
SI le passager est un Homme ET a payé un billet supérieur à 26.2$, ALORS il Meurt généralement (84 décès contre 33 survivants), sauf dans une petite tranche de prix très spécifique.

Oui, cet arbre est globalement très cohérent avec l'histoire et les théories sociologiques du naufrage, malgré quelques spécificités statistiques :

Priorité de genre : L'arbre place le sexe au sommet. C'est cohérent avec la règle "les femmes et les enfants d'abord". La survie des femmes est massivement supérieure à celle des hommes dès le premier embranchement.

Stratification sociale : La séparation par Pclass chez les femmes confirme que la richesse et la position de la cabine sur le bateau ont joué un rôle vital. Les femmes de 1ère et 2ème classe sont les mieux protégées.

Le facteur "Enfant" : On remarque que chez les hommes (Kid <= 0.5), le critère de l'enfance apparaît mais il ne suffit pas à garantir la survie dans les classes populaires (ce qui est historiquement vrai : les enfants de 3ème classe ont eu un taux de mortalité bien plus élevé que ceux de 1ère classe).

Une petite anomalie (Bruit statistique) : La règle sur le Fare > 8.04 pour les femmes de 3ème classe qui mènerait au décès peut paraître contre-intuitive (on survit moins en payant plus cher). Cela est souvent dû à des familles nombreuses de 3ème classe qui voyageaient ensemble avec des billets groupés plus chers et qui ont malheureusement péri ensemble, créant ce "bruit" dans les données.*
